```python
import pandas as pd
```

```python
df=pd.read_excel('data.xlsx')
```

```python
import datetime
```

```python
df['date']=df['time'].dt.day

df['month']=df['time'].dt.month

df['year']=df['time'].dt.year

df['day_of_week']=df['time'].dt.dayofweek
```

___________MONTH START AND MONTH END______________________

```python
my_list=[]
for date in df['date']:
    
    if date<7:
        my_list.append(1)
        
    elif date>=7 and date<=24:
        my_list.append(2)
        
    else:
        
        my_list.append(3)
        
        

        

df['part_of_month']=pd.DataFrame(my_list,columns=['part_of_month'])
```

__QUATER START AND END______________________

```python
my_list=[]
for index in df.index:
    date=df['date'].iloc[index]
    month=df['month'].iloc[index]
    
    if month in [1,4,7,10]:
        if date in [ x for x in range(1,21)]:
            my_list.append(1)
            
        else:
            
            my_list.append(2)
            
    elif month in [3,6,9,12]:
        if date in [x for x in range(11,32)]:
            
            my_list.append(3)
            
        else:
            
            my_list.append(2)
            
    else:
        
        my_list.append(2)
    
    
df['quater_start_quater_end']=pd.DataFrame(my_list)
```

_YEAR START YEAR END_____________________

```python
my_list=[]
for month in df['month']:
    
    if month<=2:
        my_list.append(1)
        
    elif month>=11:
        
        my_list.append(3)
        
    else:
        
        my_list.append(2)
        
      
df['year_start_year_end']=pd.DataFrame(my_list)
```

______WEEK START WEEK END_________________

```python
my_list=[]
for day in df['day_of_week']:
    if day in [5,6]:
        my_list.append(3)
        
    elif day in [0,1]:
        my_list.append(1)
        
    else:
        my_list.append(2)
    

df['week_start_week_end']=pd.DataFrame(my_list)
```

```python
df['hour']=df['time'].dt.hour
```

```python
my_list=[]

for hour in df['hour']:
    if hour>=0 and hour<=8:
        my_list.append(1)
        
        
    elif hour>=9 and hour<=17:
        
        my_list.append(2)
        
    else:
        my_list.append(3)
        
        
df['shifts']=pd.DataFrame(my_list)
```

_____DROPING VARIABLES______

```python
df.drop('time',axis=1,inplace=True)

df.drop(['date','day_of_week','hour'],axis=1,inplace=True)
```

```python
df=df.rename(columns={'part_of_month':'month_start_month_end'})
```

FINDING THE MISSING DATA___________________

```python
my_list=[]
for column in ['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp',
       'Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft',
       'Cyclone_Gas_Outlet_Temp', 'Cyclone_Inlet_Draft']:
    list_of_index=list(df[df[column].apply(lambda x: 1 if str(x).replace('-','').split('.')[0].isdigit() else None).isnull()].index)
    my_list.append(list_of_index)
    
    
```

```python
my_dict={}
for element in my_list:
    for x in element:
        try:
            m=my_dict[x]
            m=m+1
            my_dict[x]=m
            
            
            
        except:
            my_dict[x]=1
          
        
        
        
        
        
```

```python
one_data_missing=[]
many_data_missing=[]
for key in my_dict.keys():
    if my_dict[key]<=1:
        one_data_missing.append(key)
        
    else:
        many_data_missing.append(key)
        
```

DROPPING DATA HAVING ALOT OF DATA MISSING_________________________ 

```python
df.drop(many_data_missing,axis=0,inplace=True)
```

```python
i=1
for column in ['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp',
       'Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft',
       'Cyclone_Gas_Outlet_Temp', 'Cyclone_Inlet_Draft']:
    for index in one_data_missing:
        value=df[column].loc[index]
        digit=str(value).replace('-','').split('.')[0]
        if digit.isdigit():
            pass
        else:
            a=0
            j=0
            
            for index in range(index-4,index+4):
                
                
                value=df[column].loc[index]
                
                
                if str(value).replace('-','').split('.')[0].isdigit():
                    a=value+a
                    j=j+1
                    
                    
            new_value=a/j
            df[column].loc[index-3]=new_value
            
         
```
